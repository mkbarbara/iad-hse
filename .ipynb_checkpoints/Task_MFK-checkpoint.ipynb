# Распределение времени задержки вылета

Видим, что распределение похоже на логнормальное, но нужно проверить и другие модели:

```python
f.get_best(method = 'sumsquare_error')
```

# Изучение корреляции числа рейсов в месяц и среднего времени задержки

```python
data_more_zero = data[data['dep_delay']>0]
data_mz_dep_delay = data_more_zero.groupby('month', as_index=False).mean()[['month','dep_delay']]
data_mz_dep_delay['number_of_flights'] = data_more_zero.groupby('month', as_index=False).count()['dep_delay']
data_mz_dep_delay
```

Значения данных сильно различаются по значениям, проведем нормализацию

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
data_mz_dep_delay[['dep_delay', 'number_of_flights']] = scaler.fit_transform(data_mz_dep_delay[['dep_delay', 'number_of_flights']])
data_mz_dep_delay
```

Построим график зависимости количества полетов и времени задержки от месяца

```python
fig, ax = plt.subplots()
ax.plot(data_mz_dep_delay['month'], data_mz_dep_delay['number_of_flights'])
ax.plot(data_mz_dep_delay['month'], data_mz_dep_delay['dep_delay'])
plt.show()
```

Рассчитаем корреляцию и построим точечную диаграмму:

```python
data_mz_dep_delay['dep_delay'].corr(data_mz_dep_delay['number_of_flights'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = data_mz_dep_delay['number_of_flights'].to_numpy().reshape(data_mz_dep_delay['dep_delay'].size, 1)
y = data_mz_dep_delay['dep_delay'].to_numpy().reshape(data_mz_dep_delay['dep_delay'].size, 1)
```

```python
model = LinearRegression()
model.fit(x, y)
model.coef_
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y,  color='black')
ax.plot(x, model.predict(x), color='blue')
ax.set_xlabel('number_of_flights')
ax.set_ylabel('dep_delay')
plt.show()
```

Видим высокую корреляцию между количеством полетов и временем задержки, уравнение регрессии: y=0.65381932x

# ML

Решаем задачу регресии. target - dep_delay, признаки, которые будут использоваться - month, day, origin. Закодируем категориальный признак origin (если я правильно поняла задание, то нужно использовать не только день и аэропорт, но еще carrier, distance, и не их отдельные значения, а средние)

```python
data_ml = data
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data_ml['carrier'] = encoder.fit_transform(data_ml['carrier'])
encoder = LabelEncoder()
data_ml['origin'] = encoder.fit_transform(data_ml['origin'])
data_ml = data_ml[['month', 'day', 'origin', 'dep_delay', 'carrier', 'distance']]
data_ml
```

```python
data_ml = data_ml.groupby(['month', 'day', 'origin']).mean().reset_index()
data_ml
```

```python
data_test = data_ml[(data_ml['month']==12) & (data_ml['day']==31)]
data_train = data_ml[(data_ml['month']!=12) & (data_ml['day']!=31)]
X = data_train.drop(['dep_delay'], axis=1)
y = data_train['dep_delay']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
X_train
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(random_state = 42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_squared_error(y_test,y_pred)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_squared_error(y_test,y_pred)
```

Линейная регрессия лучше решающего дерева со стандартными параметрами

```python
print(model.predict(data_test.drop(['dep_delay'], axis=1)))
print(data_test['dep_delay'].to_list())
```
